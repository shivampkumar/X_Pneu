```python
#Step 1...Make the images uniform
import cv2
import glob
import os

# Get images
imgs = glob.glob("chest_xray\\train\\0\\"+"*.jpeg")
#imgs.extend(glob.glob('*.jpeg'))

print('Found files:')
print(imgs)

width = 1500
print('Resizing all images be %d pixels wide' % width)

folder = 'train\\0'
if not os.path.exists(folder):
    os.makedirs(folder)

# Iterate through resizing and saving
i=0
for img in imgs:
    pic = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    height =1100
    pic = cv2.resize(pic, (width, height))
    cv2.imwrite(folder + '\\' + str(i)+'.jpeg', pic)
    i= i+1
    print(i)
```
